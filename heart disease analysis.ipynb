```python
import pandas as pd

# Extract: Load the dataset
data = pd.read_csv('/content/Heart Disease data.csv')

data.shape
data.columns
```

```python
data.describe()
```

```python
data.shape
```

```python
# Transform: Data Cleaning
# Check for missing values
print(data.isnull().sum())

# Drop rows with missing values
data.dropna(inplace=True)

# Remove duplicates
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns



# Pairplot to visualize relationships between features
sns.kdeplot(data)
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.clustermap(data.corr(), annot=True, cmap='coolwarm')
plt.show()
data.to_csv('cleaned_heart_disease_data.csv', index=False)
```

```python
data = pd.read_csv('cleaned_heart_disease_data.csv')

# Bar plot for heart disease rates by age
plt.figure(figsize=(12, 6))
sns.lineplot(x='age', y='thalach', data=data)
plt.title('Heart Disease Rates by Age')
plt.xlabel('age')
plt.ylabel('Heart Disease Rate')
plt.show()

# Line plot for heart disease trends by gender
plt.figure(figsize=(12, 6))
sns.lineplot(x='age', y='thalach', hue='sex', data=data)
plt.title('Heart Disease Trends by Gender')
plt.xlabel('Age')
plt.ylabel('Heart Disease Rate')
plt.show()
```
